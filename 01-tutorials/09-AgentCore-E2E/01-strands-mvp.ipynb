# [Cross-Region inference] 

```python
# Get boto session
session = boto3.Session()
region = session.region_name
print(region)
```

```python
model = BedrockModel(
    model_id="us.deepseek.r1-v1:0",
    boto_session=session,
    streaming=False,
    temperature=0.2,
    max_tokens=1024,
)
```

```python
agent = Agent(model=model)
```

```python
q = "What is the capital of France?"
```

```python
result = agent(q)
```

```python
type(result)
```

```python
vars(result)
```

# [Single Region Model] Using a model that requires a different Region

#### DeepSeek

```python
session = boto3.Session(
    profile_name="personal",
    region_name="eu-west-2"
)
print(session.region_name)
```

```python
model_id="deepseek.v3-v1:0"
```

```python
model = BedrockModel(
    model_id=model_id,
    boto_session=session,
    streaming=False,
    temperature=0.2,
    max_tokens=1024,
)
```

```python
agent = Agent(model=model)
```

```python
q = "Describe me in two sentences what is Amazon Bedrock?"
```

```python
result = agent(q)
```
